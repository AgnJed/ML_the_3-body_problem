```python
import pandas as pd
import matplotlib.pyplot as plt 
```

```python
X_train = pd.read_csv('/kaggle/input/3vbodyproblem/mlNOVA/mlNOVA/X_train.csv')
X_test = pd.read_csv('/kaggle/input/3vbodyproblem/mlNOVA/mlNOVA/X_test.csv')
```

```python
print("Summary of the missing values in the training set\n")
print(X_train.isna().sum())

print("Summary of the missing values in the test set\n")
print(X_test.isna().sum())
```

```python
start = 0
end = 257

for i in range(5):
    trajectory = X_train[start:end]
    x1, y1, x2, y2, x3, y3 = trajectory.x_1, trajectory.y_1, trajectory.x_2, trajectory.y_2, trajectory.x_3, trajectory.y_3
    plt.plot(x1, y1, color="hotpink", label = "Body 1")
    plt.plot(x2, y2, color="blue", label = "Body 2")
    plt.plot(x3, y3, color="purple", label = "Body 3")
    plt.legend()
    plt.show()

    start += 257
    end += 257
```
